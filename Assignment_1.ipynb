```python
# importing numpy library
import numpy as np
matrix1 = np. random. randint (1,20, 15)
print ("Array")
print (matrix1)
#1. Reshape the array to 3 by 5 
matrix2=np. reshape (matrix1, (3,5)) 
print ("Reshaped Array to 3 by 5") 
print (matrix2)
# print Array shape.
print("Array Shape")
print (matrix2.shape)
#Replaced the maximum in each row by 0
index = matrix2.argmax (axis=1)
print("indexes of maximum values in array")
print (index)
print ("Array with replaced maximum value")
matrix2[0] [index [0]] = 0
matrix2[1] [index [1]] = 0
matrix2[2] [index [2]] = 0
print(matrix2)
```

```python
# importing numpy library
import numpy as np

# create numpy 2-Dymention array
matrix = np.array([[3,-2],[1,0]])

print("Printing the Original square array:\n",matrix)

# finding eigenvalues and eigenvector
x, y = np.linalg.eig(matrix)

# printing eigen values
print("Printing the Eigen Values of the given square array:\n",x)


# printing eigen vectors
print("Printing Right Eigen Vectors of the given square array:\n",y)
```

```python
# importing Numpy package
import numpy as np
  
# creating a 3X3 Numpy matrix
numpy_matrix = np.array([[0,1,2],[3,4,5]])

# Displaying the Matrix
print("Numpy Matrix is:\n",numpy_matrix)

# calculating the Trace of a matrix
trace_matrix = np.trace(numpy_matrix)

print("\nTrace of given matrix:")
print(trace_matrix)
```

```python
# importing matplotlib library
import matplotlib.pyplot as pypt
# Data to plot
program_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
# explode first slice
explode = (0.1, 0, 0, 0,0,0)
# Plotting 
wedgep={'linewidth':1,'edgecolor':'black'}
pypt.pie(popuratity, explode=explode, labels=program_languages, colors=colors,
autopct='%1.1f%%', shadow=True, wedgeprops= wedgep,startangle=140)

pypt.show()
```
